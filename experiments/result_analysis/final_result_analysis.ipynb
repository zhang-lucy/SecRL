```python
import numpy as np
from scipy.stats import pearsonr, spearmanr
from sklearn.metrics import roc_auc_score, average_precision_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

def compute_metrics(charlen, reward, normalize=False):
    """
    Compute:
      - Pearson correlation between charlen and reward
      - Spearman correlation between charlen and reward
      - ROC AUC for predicting binary reward (1 if reward == 1 else 0) using charlen as score
      - Average precision (area under PR curve) for the same

    Parameters:
      - charlen: list or array of return lengths
      - reward: list or array of rewards
      - normalize: bool, if True applies log1p transform + z-score to charlen

    Returns a dict with keys: 'pearson', 'spearman', 'roc_auc', 'average_precision'.
    """
    charlen = np.array(charlen, dtype=float)
    reward = np.array(reward, dtype=float)
    
    # Optional normalization: log1p + z-score
    if normalize:
        charlen = np.log1p(charlen)
        charlen = StandardScaler().fit_transform(charlen.reshape(-1, 1)).flatten()
    
    # 1. Pearson and Spearman
    pearson_corr, _ = pearsonr(charlen, reward)
    spearman_corr, _ = spearmanr(charlen, reward)
    
    # 2. Binary target (1 if reward == 1, else 0)
    binary_target = (reward == 1).astype(int)
    
    # 3. ROC AUC and Average Precision
    roc_auc = roc_auc_score(binary_target, charlen)
    avg_prec = average_precision_score(binary_target, charlen)
    
    return {
        'pearson': pearson_corr,
        'spearman': spearman_corr,
        'roc_auc': roc_auc,
        'average_precision': avg_prec
    }

import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

def plot_charlen_distribution(charlen, bins=30, normalize=False):
    """
    Plot a histogram of return lengths (charlen), with optional normalization
    and bin‐range labels.

    Parameters:
      - charlen: list or array of return lengths
      - bins: number of histogram bins
      - normalize: bool, if True applies log1p transform + z-score before plotting
    """
    # Prepare data
    data = np.array(charlen, dtype=float)
    if normalize:
        data = np.log1p(data)
        data = StandardScaler().fit_transform(data.reshape(-1, 1)).flatten()

    # Plot histogram and capture bin edges
    plt.figure()
    counts, bin_edges, _ = plt.hist(data, bins=bins, edgecolor='black')

    # Create human‐readable labels for each bin, e.g. "0–100", "101–200", …
    # Use integer formatting if original data are ints; fallback to two decimals otherwise
    labels = []
    for left, right in zip(bin_edges[:-1], bin_edges[1:]):
        # round edges for readability
        l = int(left) if left.is_integer() else round(left, 1)
        r = int(right) if right.is_integer() else round(right, 1)
        labels.append(f"{l}–{r}")

    # Position tick marks at the center of each bin
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    plt.xticks(bin_centers, labels, rotation=90, fontsize=8)

    # Labels and title
    xlabel = "Normalized return length" if normalize else "Return length (chars)"
    plt.xlabel(xlabel)
    plt.ylabel("Frequency")
    plt.title("Distribution of " + xlabel)
    plt.tight_layout()  # ensure labels fit
    plt.show()

def plot_reward_vs_charlen(charlen, reward, normalize=False):
    """
    Plot a scatter plot of reward versus return length, with optional normalization.
    
    Parameters:
      - charlen: list or array of return lengths
      - reward: list or array of rewards
      - normalize: bool, if True applies log1p transform + z-score to charlen
    """
    charlen_arr = np.array(charlen, dtype=float)
    if normalize:
        charlen_arr = np.log1p(charlen_arr)
        charlen_arr = StandardScaler().fit_transform(charlen_arr.reshape(-1, 1)).flatten()
    
    reward_arr = np.array(reward, dtype=float)
    
    plt.figure()
    plt.scatter(charlen_arr, reward_arr, alpha=0.5)
    xlabel = "Normalized return length" if normalize else "Return length (chars)"
    plt.xlabel(xlabel)
    plt.ylabel("Reward")
    plt.title("Reward vs " + xlabel)
    plt.show()
```

```python
def pearson_accuracy_cost(model_dict, exclude=None):
    """
    Pearson correlation (ρ) between 'accuracy' and 'cost' in a
    dictionary of model metrics, with an option to skip specified models.

    Parameters
    ----------
    model_dict : dict
        {
            "model-name": {"accuracy": float, "cost": float, ...},
            ...
        }

    exclude : iterable, optional
        List/tuple/set of model names to ignore in the computation.
        Default is None (include every model).

    Returns
    -------
    float
        Pearson correlation coefficient in [-1, 1].

    Raises
    ------
    ValueError
        • Fewer than two usable models after exclusions  
        • Constant accuracy or cost vector (zero variance)
    """
    import math

    # Normalize the exclusion list to a set for O(1) look‑ups
    exclude = set(exclude or ())

    # --- 1. Collect data, skipping excluded models -------------------
    acc, cost = [], []
    for name, metrics in model_dict.items():
        if name in exclude:
            continue
        try:
            acc.append(float(metrics["accuracy"]))
            cost.append(float(metrics["cost"]))
        except (KeyError, TypeError):
            raise ValueError(f"Model {repr(name)} lacks 'accuracy' or 'cost'")

    # --- 2. Basic checks --------------------------------------------
    if len(acc) < 2:
        raise ValueError("Need at least two models (after exclusions) to compute correlation")
    if len(set(acc)) == 1 or len(set(cost)) == 1:
        raise ValueError("Zero variance in accuracy or cost; correlation undefined")

    # --- 3. Library computation -------------------------------------
    try:
        from scipy.stats import pearsonr
        rho, _ = pearsonr(acc, cost)
        return rho
    except ModuleNotFoundError:
        import numpy as np
        return float(np.corrcoef(acc, cost)[0, 1])
    
print(f"Error rate correlation: {pearson_accuracy_cost(reward_sth_results['error_query_rate'])}")
print(f"Empty result rate correlation: {pearson_accuracy_cost(reward_sth_results['empty_result_rate'], exclude=['phi-4'])}")
print(f"Success rate correlation: {pearson_accuracy_cost(reward_sth_results['success_query_rate'])}")
print(f"Submit rate correlation: {pearson_accuracy_cost(reward_sth_results['submit_rate'], exclude=['gpt-4.1-nano', 'gpt-4o-mini'])}")
print(f"Avg return chars correlation: {pearson_accuracy_cost(reward_sth_results['avg_return_chars'])}")

```

```python
def analysis_3_trials(data: dict, verbose: bool = False):
    """
    • Aggregates rounds & tokens over every trial.
    • For each trial, picks the *last* model in `usage_summary`
      when adding prompt/completion tokens.
    • Submission / evaluation flags inspected only on the last trial
      (per earlier request).
    """
    total_len = len(data)
    total_reward = total_round = 0
    success_count = non_zero_reward_count = not_submit_count = 0
    path_count, reward_count, round_count = {}, {}, {}
    eval_error_count = fail_to_run_count = 0
    total_prompt_tokens = total_completion_tokens = total_cost = 0
    empty_result_count = error_query_count = query_count = 0  # placeholders

    for q in data:
        p_len = len(q["question_dict"]["shortest_alert_path"])
        path_count.setdefault(p_len, 0)
        reward_count.setdefault(p_len, 0)
        round_count.setdefault(p_len, 0)

        # reward bookkeeping (once per question)
        reward = q["reward"]
        total_reward += reward
        reward_count[p_len] += reward
        if reward > 0:
            non_zero_reward_count += 1
        if reward == 1:
            success_count += 1
        path_count[p_len] += 1

        # ---- iterate over ALL trials ------------------------------------
        for trial in q["trials"].values():
            msgs = trial.get("messages", [])
            tmp_round = max((len(msgs) - 1) // 2, 0)
            total_round += tmp_round
            round_count[p_len] += tmp_round

            # === NEW token logic: use *last* model =======================
            usage_summary = trial.get("usage_summary", {})
            if usage_summary:                                   # guard empty dict
                try:
                    mdl = list(usage_summary.keys())[-1]        # last‑listed model
                    total_prompt_tokens     += usage_summary[mdl].get("prompt_tokens", 0)
                    total_completion_tokens += usage_summary[mdl].get("completion_tokens", 0)
                except Exception as e:
                    print(f"[usage error] q nodes={q['nodes']} trial={mdl}: {e}")

        # ---- submission / evaluation checks on final trial -------------
        last_trial = list(q["trials"].values())[-1]
        if not last_trial["info"].get("submit"):
            not_submit_count += 1
        elif not (last_trial["info"].get("is_json_success", True)
                  and last_trial["info"].get("is_reflect_success", True)):
            print(f"[eval error] q nodes={q['nodes']}: {last_trial['info']}")
            eval_error_count += 1

    # ---- optional console summary --------------------------------------
    if verbose:
        print(f"Average reward: {total_reward / total_len:.6f}")
        print(f"Average round:  {total_round  / total_len:.6f}")

    return {
        "total_len": total_len,
        "total_reward": total_reward,
        "success_count": success_count,
        "non_zero_reward_count": non_zero_reward_count,
        "not_submit_count": not_submit_count,
        "total_round": total_round,
        "total_prompt_tokens": total_prompt_tokens,
        "total_completion_tokens": total_completion_tokens,
        "total_cost": total_cost,   # still placeholder
        "empty_result_count": empty_result_count,
        "error_query_count": error_query_count,
        "query_count": query_count,
        "eval_error_count": eval_error_count,
        "fail_to_run_count": fail_to_run_count,
    }

def get_avg_accuracy_and_cost_3_trial(log_path, file_folder, model_name, version="v1", round_cut=-1):
    d = 1_000_000  # price is per 1M tokens
    prices = {
        "gpt-4o": (2.5 / d, 10 / d),
        "gpt-4o-mini": (0.15 / d, 0.6 / d),
        "o1-mini": (1.1 / d, 4.4 / d),
        "o1": (15 / d, 7.5 / d),
        "o3-mini": (1.1 / d, 4.4 / d),
        "deepseek-r1": (0.55 / d, 0.27 / d),
        "phi4": (0.07/ d, 0.14/ d),
    }

    prompt_price, completion_price = prices[model_name]
    file_template = f"{log_path}/{file_folder}" + "/agent_incident_{0}.json"

    total_count = 0
    total_reward = 0
    total_prompt_tokens = 0
    total_completion_tokens = 0
    total_round = 0

    incidents = [5, 34, 38, 39, 55, 134, 166, 322]
    for i in incidents:
        with open(file_template.format(i), "r") as f:
            data = json.load(f)

        if version == "v2":
            result = analysis_3_trials(data)
        else:
            raise ValueError("Unknown version: {}".format(version))

        total_count += result['total_len']
        total_reward += result['total_reward']
        total_prompt_tokens += result['total_prompt_tokens']
        total_completion_tokens += result['total_completion_tokens']
        total_round += result['total_round']

    avg_accuracy = round(total_reward / total_count, 3)
    avg_cost = round(
        (total_prompt_tokens * prompt_price + total_completion_tokens * completion_price)
        / total_count,
        3,
    )
    avg_round = round(total_round / total_count, 1)

    return avg_accuracy, avg_cost, avg_round

tab2_3_trial = {
  "strategy-gpt-4o": ("PromptSauceAgent_gpt-4o_c345_alert_level_t0_s15_trial3", "v2"),
  "strategy-reflect-gpt-4o": ("PrologmptSauceReflexionAgent_gpt-4o_c234_alert_level_t0_s15_trial3", "v2"),
  "react-gpt-4o": ("ReactAgent_gpt-4o_c121_alert_level_t0_s15_trial3", "v2"),
  "react-reflect-gpt-4o": ("ReActReflexionAgent_gpt-4o_c378_alert_level_t0_s15_trial3", "v2"),

  "strategy-4o-mini": ("PromptSauceAgent_4o-mini_c34_alert_level_t0_s15_trial3", "v2"),
  "strategy-reflect-4o-mini": ("PromptSauceReflexionAgent_4o-mini_c234_alert_level_t0_s15_trial3", "v2"),
  "react-reflexion-4o-mini": ("ReActReflexionAgent_4o-mini_c378_alert_level_t0_s15_trial3", "v2"),
  "react-4o-mini": ("ReactAgent_4o-mini_c124_alert_level_t0_s15_trial3", "v2"),

  "strategy-o3-mini": ("PromptSauceAgent_o3-mini_c577_alert_level_t0_s15_trial3", "v2"),
  "strategy-reflect-o3-mini": ("PromptSauceReflexionAgent_o3-mini_c677_alert_level_t0_s15_trial3", "v2"),
  "react-o3-mini": ("ReactAgent_o3-mini_c476_alert_level_t0_s15_trial3", "v2"),
  "react-reflexion-o3-mini": ("ReActReflexionAgent_o3-mini_c877_alert_level_t0_s15_trial3", "v2"),
  



}

log_path = "../secgym/final_results"
for name, b in tab2_3_trial.items():
    if type(b) == tuple:
        file_folder, version = b
    else:
        file_folder = b
        version = "v1"
    if "gpt-4o" in name:
        model_name = "gpt-4o"
    elif "4o-mini" in name:
        model_name = "gpt-4o-mini"
    elif "o3-mini" in name:
        model_name = "o3-mini"
    else:
        raise ValueError(f"Unknown model name: {name}")
    
    print(name, version)
    avg_accuracy, avg_cost, avg_round = get_avg_accuracy_and_cost_3_trial(log_path, file_folder, model_name, version=version)
    print(f"Avg Accuracy: {avg_accuracy}, Avg Cost: {avg_cost}", 
          f"Avg Round: {avg_round}")
    print(f"& {avg_accuracy} & {avg_round} & {avg_cost} ")
    print("===================")
```

```python
def analysis_3_trials(data: dict, verbose: bool = False, trial_num=None):
    """
    • Aggregates rounds & tokens over every trial.
    • For each trial, picks the *last* model in `usage_summary`
      when adding prompt/completion tokens.
    • Submission / evaluation flags inspected only on the last trial
      (per earlier request).
    """
    total_len = len(data)
    total_reward = total_round = 0
    success_count = non_zero_reward_count = not_submit_count = 0
    path_count, reward_count, round_count = {}, {}, {}
    eval_error_count = fail_to_run_count = 0
    total_prompt_tokens = total_completion_tokens = total_cost = 0
    empty_result_count = error_query_count = query_count = 0  # placeholders

    for q in data:
        p_len = len(q["question_dict"]["shortest_alert_path"])
        path_count.setdefault(p_len, 0)
        reward_count.setdefault(p_len, 0)
        round_count.setdefault(p_len, 0)

        if trial_num is None:
            trial_num = len(q["trials"])
        last_trial_num = min(trial_num, len(q["trials"]))
        # reward bookkeeping (once per question)
        reward = q['trials'][str(last_trial_num - 1)]["reward"]
        total_reward += reward
        reward_count[p_len] += reward
        if reward > 0:
            non_zero_reward_count += 1
        if reward == 1:
            success_count += 1
        path_count[p_len] += 1

        # ---- iterate over ALL trials ------------------------------------

        
        for n in range(last_trial_num):
            trial = q["trials"][str(n)]
            msgs = trial.get("messages", [])
            tmp_round = max((len(msgs) - 1) // 2, 0)
            total_round += tmp_round
            round_count[p_len] += tmp_round

            # === NEW token logic: use *last* model =======================
            usage_summary = trial.get("usage_summary", {})
            if usage_summary:                                   # guard empty dict
                try:
                    mdl = list(usage_summary.keys())[-1]        # last‑listed model
                    total_prompt_tokens     += usage_summary[mdl].get("prompt_tokens", 0)
                    total_completion_tokens += usage_summary[mdl].get("completion_tokens", 0)
                except Exception as e:
                    print(f"[usage error] q nodes={q['nodes']} trial={mdl}: {e}")
        # ---- submission / evaluation checks on final trial -------------
        last_trial = list(q["trials"].values())[last_trial_num - 1]

        last_trial = list(q["trials"].values())[-1]
        if not last_trial["info"].get("submit"):
            not_submit_count += 1
        elif not (last_trial["info"].get("is_json_success", True)
                  and last_trial["info"].get("is_reflect_success", True)):
            print(f"[eval error] q nodes={q['nodes']}: {last_trial['info']}")
            eval_error_count += 1

    # ---- optional console summary --------------------------------------
    if verbose:
        print(f"Average reward: {total_reward / total_len:.6f}")
        print(f"Average round:  {total_round  / total_len:.6f}")

    return {
        "total_len": total_len,
        "total_reward": total_reward,
        "success_count": success_count,
        "non_zero_reward_count": non_zero_reward_count,
        "not_submit_count": not_submit_count,
        "total_round": total_round,
        "total_prompt_tokens": total_prompt_tokens,
        "total_completion_tokens": total_completion_tokens,
        "total_cost": total_cost,   # still placeholder
        "empty_result_count": empty_result_count,
        "error_query_count": error_query_count,
        "query_count": query_count,
        "eval_error_count": eval_error_count,
        "fail_to_run_count": fail_to_run_count,
    }
def get_avg_accuracy_and_rounds_per_trial(log_path, file_folder, version="v1", trial_num=None):
    """
    Get average accuracy and rounds for a specific trial number or all trials.
    
    Args:
        log_path: Path to log directory
        file_folder: Folder containing agent results
        version: Version of analysis function to use
        trial_num: Specific trial number to analyze (1-10), or None for all trials
        
    Returns:
        tuple: (avg_accuracy, avg_round)
    """
    file_template = f"{log_path}/{file_folder}" + "/agent_incident_{0}.json"

    total_count = 0
    total_reward = 0
    total_round = 0

    incidents = [5, 34, 38, 39, 55, 134, 166, 322]
    for i in incidents:
        with open(file_template.format(i), "r") as f:
            data = json.load(f)

        if version == "v2":
            result = analysis_3_trials(data, trial_num=trial_num)
        else:
            raise ValueError("Unknown version: {}".format(version))

        total_count += result['total_len']
        total_reward += result['total_reward']
        total_round += result['total_round']

    avg_accuracy = round(total_reward / total_count, 3)
    avg_round = round(total_round / total_count, 1)

    return avg_accuracy, avg_round


file_path = "BaselineAgent_gpt-4o_c8801_alert_level_t0_s15_trial10"
rewards = []
rounds = []
for i in range(1, 11):
    avg_accuracy, avg_round = get_avg_accuracy_and_rounds_per_trial(log_path, file_path, version="v2", trial_num=i)
    rewards.append(avg_accuracy)
    rounds.append(avg_round)
    print(f"Trial {i}: Avg Accuracy: {avg_accuracy}, Avg Round: {avg_round}")
```
